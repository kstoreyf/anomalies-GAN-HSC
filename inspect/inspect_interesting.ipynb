```python
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
hdul = fits.open('ADP.2015-10-15T12_53_33.873.fits')
```

```python
data = hdul[1].data
```

```python
data
```

```python
wl = data[0][0]
flux = data[0][1]
```

```python
plt.figure(figsize=(12,8))
plt.plot(wl, flux/1e-17, color='black')
plt.xlabel(r"wavelength ($\AA$)")
#plt.ylabel(r"flux (10$^{-17}$ erg/s/cm$^2$/$\AA$)")
plt.ylabel(r"flux / 10$^{-17}$")

#plt.yscale('log')
```
